```python
import synth_training_set as sts
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten
import itertools
import os
import datetime


TSX=sts.TEXT_SHAPE[1]
TSY=sts.TEXT_SHAPE[0]
batch_size=50

def generate_batch(size):
    l=[]
    text_present=[]
    i=0
    for im, ch, istext in itertools.islice(sts.gen_training_im(), size):        
        l+=[im]
        text_present+=[int(istext)]
        name="./batch/{:08d}_{}_{}.png".format(i,ch,"1" if istext else "0")  
        #print(name)
        #cv2.imwrite(name,im*255.)
        i+=1
    return np.array(l), np.array(text_present)

def get_model():
    model=Sequential()
    model.add(Conv2D(16, kernel_size=3, kernel_initializer='glorot_uniform',activation='relu',input_shape=(TSY,TSX,1)))
    model.add(MaxPool2D(pool_size=(2,2)))
    model.add(Conv2D(32, kernel_size=3, activation='relu'))
#    model.add(MaxPool2D(pool_size=(2,2)))
    model.add(Conv2D(64, kernel_size=3, activation='relu'))    
#    model.add(Conv2D(64, kernel_size=5,activation='relu')) 
    model.add(Flatten())
    model.add(Dense(32,input_dim=TSX*TSY,activation=None))
    model.add(Dense(1,activation='sigmoid'))
    
    return model

```

```python
def get_training_set():
    while(True):
        X_train,Y_train=generate_batch(1000)
        #print(X_train.shape)
        #print(Y_train.shape)
        #X_test,Y_test=generate_batch(int(batch_size*0.2))
        yield X_train.reshape(X_train.shape[0],X_train.shape[1],X_train.shape[2],1),Y_train.reshape(Y_train.shape[0], 1)
        #yield X_train,Y_train

def get_validation_set():
    while(True):
        X_train,Y_train=generate_batch(1000)
        print(X_train.shape)
        print(Y_train.shape)
        #X_test,Y_test=generate_batch(int(batch_size*0.2))
        return X_train.reshape(X_train.shape[0],X_train.shape[1],X_train.shape[2],1),Y_train.reshape(Y_train.shape[0], 1)
                
```

```python
#X_train,Y_train,X_test,Y_test=get_training_set(batch_size)
#X_tr=X_train.reshape(X_train.shape[0],X_train.shape[1],X_train.shape[2],1)
#X_tt=X_test.reshape(X_test.shape[0],X_test.shape[1],X_test.shape[2],1)
model=get_model()
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

#print(model.summary())
    #for layer in model.layers:
    #   print("Input shape: "+str(layer.input_shape)+". Output shape: "+str(layer.output_shape))
    
X_val,Y_val=get_validation_set()
model.fit_generator(get_training_set(),epochs=500,steps_per_epoch=10)#,callbacks=[tensorboard_callback], validation_data=(X_val,Y_val))#,validation_steps=10)

#model.fit(X_tr,Y_train,validation_data=(X_tt,Y_test),epochs=10)
```
